```python
import numpy as np
import matplotlib.pyplot as plt

# Generarea datelor astfel încât y = 2x + 20
x = np.array([1, 3, 5, 2.5, 2, 4])
y = 2 * x + 2

a = 0.1  # Coeficientul inițial
b = 1  # Bias inițial
learning_rate = 0.0375

def mse(a, b, x, y):
    y_pred = a * x + b
    return np.mean((y_pred - y) ** 2)

def gradient_steps(a, b, learning_rate, x, y):
    for i in range(len(x)):
        x_i, y_i = x[i], y[i]
        grad_a = 2 * (a * x_i + b - y_i) * x_i  # Gradient pentru a
        grad_b = 2 * (a * x_i + b - y_i)  # Gradient pentru b
        
        a -= learning_rate * grad_a  # Actualizare a
        b -= learning_rate * grad_b  # Actualizare b

        plt.figure(figsize=(14, 3))
        plt.subplot(1, 2, 1)
        plt.scatter(x, y, color='red', label='Puncte de date')
        plt.plot(x, a * x + b, label=f'y = {a:.4f}x + {b:.4f}', color='blue')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.legend()
        plt.title(f'Iterația {i+1}: Modelul liniar')

        a_values = np.linspace(-1, 5, 100)
        mse_values = [mse(a_val, b, x, y) for a_val in a_values]
        plt.subplot(1, 2, 2)
        plt.plot(a_values, mse_values, label='MSE Curve', color='purple')
        plt.scatter(a, mse(a, b, x, y), color='black', s=100, label='Gradient step')
        plt.xlabel('a')
        plt.ylabel('MSE')
        plt.legend()
        plt.title('Evoluția erorii')
        
        print(f'Error: {mse(a, b, x, y):.4f}, Gradient sign (a, b): ({np.sign(grad_a)}, {np.sign(grad_b)}), a: {a + learning_rate * grad_a} -> {a}, b: {b + learning_rate * grad_b} -> {b}')
        
        plt.show()
    
    return a, b
```

```python
a, b = gradient_steps(a, b, learning_rate, x, y)
```

```python
for _ in range(10):
    a, b = gradient_steps(a, b, learning_rate, x, y)
```
