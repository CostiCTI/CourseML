```python
import numpy as np
import matplotlib.pyplot as plt

# Datele furnizate
x = np.array([10, 30, 50, 25, 20, 40])
y = np.array([20, 60, 100, 50, 40, 80])

a = 0.1 
learning_rate = 0.00025

def mse(a, x, y):
    y_pred = a * x
    return np.mean((y_pred - y) ** 2)

def gradient_steps(a, learning_rate, x, y):
    for i in range(len(x)):
        x_i, y_i = x[i], y[i]
        grad = 2 * (a * x_i - y_i) * x_i 
        a -= learning_rate * grad 

        plt.figure(figsize=(14, 3))
        plt.subplot(1, 2, 1)
        plt.scatter(x, y, color='red', label='Puncte de date')
        plt.plot(x, a * x, label=f'y = {a:.4f}x', color='blue')
        #plt.scatter(x_i, y_i, color='green', s=100, label='Punct curent')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.legend()
        plt.title(f'Iterația {i+1}: Modelul liniar')

        a_values = np.linspace(-1, 5, 100)
        mse_values = [mse(a_val, x, y) for a_val in a_values]
        plt.subplot(1, 2, 2)
        plt.plot(a_values, mse_values, label='MSE Curve', color='purple')
        plt.scatter(a, mse(a, x, y), color='black', s=100, label='Gradient step')
        plt.xlabel('a')
        plt.ylabel('MSE')
        plt.legend()
        plt.title('Evoluția erorii')
        
        print ('Error:', mse(a, x, y), 'Gradient sign:', np.sign(grad), 'a:', a + learning_rate * grad, '->', a) #grad, x_i, y_i, a

        plt.show()
    
    return a
```

```python
a = gradient_steps(a, learning_rate, x, y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
```

```python
a = 2
b = 3
beta = 5

x1 = np.array([1, 2, 3, 4])
x2 = np.array([5, 6, 7, 8])

y = a * x1 + b * x2 + beta

X = np.column_stack((x1, x2))
model = LinearRegression()
model.fit(X, y)

fig = plt.figure(figsize=(14, 14))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x1, x2, y, color='b', label='Data points')


x1_grid, x2_grid = np.meshgrid(np.linspace(0, 5, 10), np.linspace(4, 9, 10))
y_grid = model.predict(np.column_stack((x1_grid.ravel(), x2_grid.ravel()))).reshape(x1_grid.shape)


ax.plot_surface(x1_grid, x2_grid, y_grid, alpha=0.5, color='r', rstride=100, cstride=100, label='Regression plane')


ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
ax.set_title('3D Linear Regression: y = ax1 + bx2 + beta')


plt.show()
```
